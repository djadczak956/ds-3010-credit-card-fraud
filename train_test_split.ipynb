```python
import pandas as pd

# load all data
all_data = pd.read_csv("card_transdata.csv")
```

```python
from sklearn.model_selection import train_test_split

# split data into training (90%) and test (10%)
training_data, test_data = train_test_split(
    all_data,
    test_size=0.1,
    random_state=42,
    shuffle=True
)

print(f"Length of all data: {len(all_data)}")
print(f"Length of training data: {len(training_data)}")
print(f"Length of testing data: {len(test_data)}")
```

```python
# save to csv's
training_data.to_csv("training_data.csv", index=False)
test_data.to_csv("test_data.csv", index=False)
```
